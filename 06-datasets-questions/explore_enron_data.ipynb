```python
quantified_salary = 0
known_emails = 0
for key,value in enron_data.items():
    if enron_data[key]['salary'] != 'NaN':
        quantified_salary += 1
    if enron_data[key]['email_address'] != 'NaN':
        known_emails += 1
print(quantified_salary,known_emails)
```

```python
nan_total_payments = 0
for key,value in enron_data.items():
    if enron_data[key]['total_payments'] == 'NaN':
        nan_total_payments += 1
nan_total_payments/len(enron_data.keys())
```

```python
poi_nan_total_payments = 0
for key,value in enron_data.items():
    if enron_data[key]['total_payments'] == 'NaN' and enron_data[key]['poi'] == True:
        poi_nan_total_payments += 1
poi_nan_total_payments/len(enron_data.keys())
```
